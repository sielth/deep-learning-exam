```python
# Load dataframe
mood_df = pd.read_csv('/content/drive/MyDrive/deeplearning-intro/project/mood_data.csv')
mood_df.columns=['name', 'album', 'artist', 'id', 'release_date', 'popularity', 'length', 'danceability', 'acousticness', 'energy', 'instrumentalness', 'liveness', 'valence', 'loudness', 'speechiness', 'tempo', 'key', 'time_signature', 'mood', 'lyrics']

mood_df
```

```python
def clean_lyrics(lyrics, title):

  if len(lyrics) > 5000:
    return 'nan'
  lyrics = re.sub(r'\[.*?\]', '', lyrics)  # remove content between square brackets
  if "Contributors" in lyrics:
    lyrics = re.sub(r'\d+ Contributors', '', lyrics) # remove "Contributors"
  if "Lyrics" in lyrics:
    lyrics = re.sub(r'Lyrics', '', lyrics)  # remove "Lyrics"
  if "Embed" in lyrics:
    lyrics = re.sub(r'Embed', '', lyrics)  # remove "Embed"
  if "\n" in lyrics:
    lyrics = re.sub(r'\n', ' ', lyrics) # remove "\n"
  match = re.search(r'\b{}\b'.format(re.escape(title)), lyrics, re.IGNORECASE) # search for the title
  if match:
    lyrics = lyrics[match.start()+len(match.group()):]  # remove everything from the start to the first occurrence of the title

  # remove number at end of lyrics
  lyrics = re.sub(r'\s*\d{1,3}$', '', lyrics)

  return lyrics.strip()  # remove leading/trailing whitespaces
```

```python
title = mood_df['name']

mood_df['lyrics'] = mood_df.apply(lambda x: clean_lyrics(str(x['lyrics']), str(x['name'])), axis=1)
```
